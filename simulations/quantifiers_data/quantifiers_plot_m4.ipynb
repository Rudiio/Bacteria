# Means calculations

After calculating the quantifiers with the corresponding jupyter notebook, the first step is to calculate the means for the serie of simulation. 

The following code calculate the means for a certain serie.

```python
# name of the serie of  file
file = "simuc1"

# all_folders= ["simuc1","simuc2","simuc3","simuc4","simuc5","simuc6","simuc8","simuc9","simuc10",
# "simuc11","simuc12","simuc13","simuc14","simuc15","simuc16","simuc17","simuc18"]

# for file in all_folders:


#aspect ratio
all_files = glob.glob("./aspect_ratio/"+ file +"/*.npy")
arrays = [np.load(all_files[i]) for i in range(len(all_files))]
S = arrays[0]
for i in range(1,len(arrays)):
    S += arrays[i]
S/=len(arrays)
np.save("./aspect_ratio/"+ file +"/"+ file + "_aspect_ratio_mean.npy",S)

#density
all_files = glob.glob("./density/"+ file +"/*.npy")
arrays = [np.load(all_files[i]) for i in range(len(all_files))]
S = arrays[0]
for i in range(1,len(arrays)):
    S += arrays[i]
S/=len(arrays)
np.save("./density/"+ file +"/"+ file + "_density_mean.npy",S)

#length
all_files = glob.glob("./length/"+ file +"/*.npy")
arrays = [np.load(all_files[i]) for i in range(len(all_files))]
S = arrays[0]
for i in range(1,len(arrays)):
    S += arrays[i]
S/=len(arrays)
np.save("./length/"+ file +"/"+ file + "_length_mean.npy",S)

#Number
all_files = glob.glob("./N/"+ file +"/*.npy")
arrays = [np.load(all_files[i]) for i in range(len(all_files))]
S = arrays[0]
for i in range(1,len(arrays)):
    S += arrays[i]
S/=len(arrays)
np.save("./N/"+ file +"/"+ file + "_N_mean.npy",S)

#El
all_files = glob.glob("./energies/"+ file +"/El/*.npy")
arrays = [np.load(all_files[i]) for i in range(len(all_files))]
S = arrays[0]
for i in range(1,len(arrays)):
    S += arrays[i]
S/=len(arrays)
np.save("./energies/"+ file +"/"+ file +"_El_mean",S)

#Et
all_files = glob.glob("./energies/"+ file +"/Et/*.npy")
arrays = [np.load(all_files[i]) for i in range(len(all_files))]
S = arrays[0]
for i in range(1,len(arrays)):
    S += arrays[i]
S/=len(arrays)
np.save("./energies/"+ file +"/"+ file +"_Et_mean",S)
```
