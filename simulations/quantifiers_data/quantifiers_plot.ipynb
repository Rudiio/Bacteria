```python
import numpy as np
import matplotlib.pyplot as plt
import glob
```

# Means calculations

```python
file = "simuc1"

#aspect ratio
all_files = glob.glob("./aspect_ratio/"+ file +"/*.npy")
arrays = [np.load(all_files[i]) for i in range(len(all_files))]
S = arrays[0]
for i in range(1,len(arrays)):
    S += arrays[i]
S/=len(arrays)
np.save("./aspect_ratio/"+ file +"/"+ file + "_aspect_ratio_mean.npy",S)

#density
all_files = glob.glob("./density/"+ file +"/*.npy")
arrays = [np.load(all_files[i]) for i in range(len(all_files))]
S = arrays[0]
for i in range(1,len(arrays)):
    S += arrays[i]
S/=len(arrays)
np.save("./density/"+ file +"/"+ file + "_density_mean.npy",S)

#length
all_files = glob.glob("./length/"+ file +"/*.npy")
arrays = [np.load(all_files[i]) for i in range(len(all_files))]
S = arrays[0]
for i in range(1,len(arrays)):
    S += arrays[i]
S/=len(arrays)
np.save("./length/"+ file +"/"+ file + "_length_mean.npy",S)

#El
all_files = glob.glob("./energies/"+ file +"/El/*.npy")
arrays = [np.load(all_files[i]) for i in range(len(all_files))]
S = arrays[0]
for i in range(1,len(arrays)):
    S += arrays[i]
S/=len(arrays)
np.save("./energies/"+ file +"/"+ file +"_El_mean",S)

#Et
all_files = glob.glob("./energies/"+ file +"/Et/*.npy")
arrays = [np.load(all_files[i]) for i in range(len(all_files))]
S = arrays[0]
for i in range(1,len(arrays)):
    S += arrays[i]
S/=len(arrays)
np.save("./energies/"+ file +"/"+ file +"_Et_mean",S)

#Ec
all_files = glob.glob("./energies/"+ file +"/El/*.npy")
arrays = [np.load(all_files[i]) for i in range(len(all_files))]
S = arrays[0]
for i in range(1,len(arrays)):
    S += arrays[i]
S/=len(arrays)
np.save("./energies/"+ file +"/"+ file +"_Ec_mean",S)
```
