# Variating ks

```python
import numpy as np
import matplotlib.pyplot as plt

L = [i for i in range(3,200,3)]
L.append(200)

# Simuc 1
L = np.array(L)
ar1 = np.load("./aspect_ratio/simuc1_aspect_ratio.npy")
d1 = np.load("./density/simuc1_density.npy")
s1 = np.load("./length/simuc1_length.npy")

# Simuc 2
ar2 = np.load("./aspect_ratio/simuc2_aspect_ratio.npy")
d2 = np.load("./density/simuc2_density.npy")
s2 = np.load("./length/simuc2_length.npy")

# Simuc 3
ar3 = np.load("./aspect_ratio/simuc3_aspect_ratio.npy")
d3 = np.load("./density/simuc3_density.npy")
s3 = np.load("./length/simuc3_length.npy")

#Simuc 8
ar8 = np.load("./aspect_ratio/simuc8_aspect_ratio.npy")
d8 = np.load("./density/simuc8_density.npy")
s8 = np.load("./length/simuc8_length.npy")

plt.figure(figsize=[10,8])
plt.plot(L,ar1,"x-",label="$m4 k_s$ = 10")
plt.plot(L,ar2,"x-",label="$m4 k_s$ = 1")
plt.plot(L,ar3,"x-",label="$m4 k_s$ = 0.1")
# plt.plot(L,ar8,"x-",label="$m5 k_s$ = 10")
plt.title("Aspect ratio as a function of time")
plt.xlabel("time (in min)")
plt.ylabel(r"$\alpha_r$")
plt.legend()
plt.show()


plt.figure(figsize=[10,8])
plt.plot(L,d1,"x-",label="$m4 k_s$ = 10")
plt.plot(L,d2,"x-",label="$m4 k_s$ = 1")
plt.plot(L,d3,"x-",label="$m4 k_s$ = 0.1")
# plt.plot(L,d8,"x-",label="$m5 k_s$ = 10")
plt.title("density as a function of time")
plt.xlabel("time (in min)")
plt.ylabel(r"$\delta$")
plt.legend()
plt.show()

plt.figure(figsize=[10,8])
plt.plot(L,s1,"x-",label="$m4 k_s$ = 10")
plt.plot(L,s2,"x-",label="$m4 k_s$ = 1")
plt.plot(L,s3,"x-",label="$m4 k_s$ = 0.1")
# plt.plot(L,s8,"x-",label="$m5 k_s$ = 10")
plt.title("length quantifier as a function of time")
plt.xlabel("time (in min)")
plt.ylabel(r"$s$")
plt.legend()
plt.show()
```
