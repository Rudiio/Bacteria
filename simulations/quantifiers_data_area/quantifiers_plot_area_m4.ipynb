# Means calculations

After calculating the quantifiers with the corresponding jupyter notebook, the first step is to calculate the means for the serie of simulation. 

The following code calculate the means for a certain serie.

```python
file = "simuc12"

#aspect ratio
all_files = glob.glob("./aspect_ratio/"+ file +"/*.npy")
arrays = [np.load(all_files[i]) for i in range(len(all_files))]
S = arrays[0]
for i in range(1,len(arrays)):
    S += arrays[i]
S/=len(arrays)

if "./aspect_ratio/"+ file +"/"+ file + "_aspect_ratio_mean4.npy" in all_files:
    os.remove("./aspect_ratio/"+ file +"/"+ file + "_aspect_ratio_mean4.npy")

np.save("./aspect_ratio/"+ file +"/"+ file + "_aspect_ratio_mean4.npy",S)

#density
all_files = glob.glob("./density/"+ file +"/*.npy")
arrays = [np.load(all_files[i]) for i in range(len(all_files))]
S = arrays[0]
for i in range(1,len(arrays)):
    S += arrays[i]
S/=len(arrays)

if "./density/"+ file +"/"+ file + "_density_mean4.npy" in all_files:
    os.remove("./density/"+ file +"/"+ file + "_density_mean4.npy")
np.save("./density/"+ file +"/"+ file + "_density_mean4.npy",S)

#length
all_files = glob.glob("./length/"+ file +"/*.npy")
arrays = [np.load(all_files[i]) for i in range(len(all_files))]
S = arrays[0]
for i in range(1,len(arrays)):
    S += arrays[i]
S/=len(arrays)

if "./length/"+ file +"/"+ file + "_length_mean4.npy" in all_files: 
    os.remove("./length/"+ file +"/"+ file + "_length_mean4.npy")
np.save("./length/"+ file +"/"+ file + "_length_mean4.npy",S)

#Number
all_files = glob.glob("./N/"+ file +"/*.npy")
arrays = [np.load(all_files[i]) for i in range(len(all_files))]
S = np.zeros(arrays[0].shape[0])
for i in range(0,len(arrays)):
    S += arrays[i]
avg = []
for i in range(S.shape[0]):
    avg.append(S[i]/len(arrays))

if "./N/"+ file +"/"+ file + "_N_mean4.npy" in all_files:
    os.remove("./N/"+ file +"/"+ file + "_N_mean4.npy")
np.save("./N/"+ file +"/"+ file + "_N_mean4.npy",np.array(avg))

#El
all_files = glob.glob("./energies/"+ file +"/El/*.npy")
arrays = [np.load(all_files[i]) for i in range(len(all_files))]
S = arrays[0]
for i in range(1,len(arrays)):
     S += arrays[i]
S/=len(arrays)

if "./energies/"+ file +"/"+ file +"_El_mean4" in all_files: 
    os.remove("./energies/"+ file +"/"+ file +"_El_mean4")
np.save("./energies/"+ file +"/"+ file +"_El_mean4",S)

#Et
all_files = glob.glob("./energies/"+ file +"/Et/*.npy")
arrays = [np.load(all_files[i]) for i in range(len(all_files))]
S = arrays[0]
for i in range(1,len(arrays)):
    S += arrays[i]
S/=len(arrays)

if "./energies/"+ file +"/"+ file +"_Et_mean4" in all_files:
    os.remove("./energies/"+ file +"/"+ file +"_Et_mean4")
np.save("./energies/"+ file +"/"+ file +"_Et_mean4",S)
```
